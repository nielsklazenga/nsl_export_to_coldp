```python
import pandas as pd
import numpy as np
```

```python
df_taxon = pd.read_csv('coldp/bryophytes/taxon.tsv', sep='\t')
df_taxon['taxonConceptID'] = df_taxon['ID']
df_taxon['acceptedNameUsageID'] = df_taxon['ID']
df_taxon['taxonomicStatus'] = 'accepted'
df_taxon.rename(columns={'ID': 'taxonID', 'nameID': 'scientificNameID', 'parentID': 'parentNameUsageID', 'remarks': 'taxonRemarks', 'accordingToID': 'nameAccordingToID'}, inplace=True)
df_taxon = df_taxon[['taxonID',
   'taxonConceptID',
    'scientificNameID',
    'nameAccordingToID',
    'taxonomicStatus',
    'acceptedNameUsageID',
    'parentNameUsageID',
    'taxonRemarks' ]]
df_taxon
```

```python
df_synonym = pd.read_csv('coldp/bryophytes/synonym.tsv', sep='\t')
df_synonym['acceptedNameUsageID'] = df_synonym['taxonID']
df_synonym['nameAccordingToID'] = np.NaN
df_synonym['parentNameUsageID'] = np.NaN
df_synonym['taxonomicStatus'] = 'synonym'
df_synonym.rename(columns={'taxonID': 'taxonConceptID', 'nameID': 'scientificNameID', 'remarks': 'taxonRemarks'}, inplace=True)
df_synonym['taxonID'] = df_synonym['scientificNameID']
df_synonym = df_synonym[['taxonID',
   'taxonConceptID',
    'scientificNameID',
    'nameAccordingToID',
    'taxonomicStatus',
    'acceptedNameUsageID',
    'parentNameUsageID',
    'taxonRemarks' ]]
df_synonym
```

```python
df_synonym = df_synonym.merge(df_taxon[['taxonID', 'parentNameUsageID']], how='left', left_on='acceptedNameUsageID', right_on='taxonID')\
    .rename(columns={'taxonID_x': 'taxonID', 'parentNameUsageID_y': 'parentNameUsageID'})[['taxonID',
   'taxonConceptID',
    'scientificNameID',
    'nameAccordingToID',
    'taxonomicStatus',
    'acceptedNameUsageID',
    'parentNameUsageID',
    'taxonRemarks' ]]
df_synonym
```

```python
# concatenate taxon and synonym dataframes
df_nameusage = pd.concat([df_taxon, df_synonym])
df_nameusage
```

```python
# get scientific name
df_name = pd.read_csv('coldp/bryophytes/name.tsv', sep='\t')

df_nameusage = df_nameusage.merge(df_name[['ID', 'scientificName', 'authorship']], how='left', left_on='scientificNameID', right_on='ID')[['taxonID',
        'taxonConceptID',
            'scientificNameID',
            'nameAccordingToID',
            'taxonomicStatus',
            'acceptedNameUsageID',
            'parentNameUsageID',
            'taxonRemarks',
            'scientificName',
            'authorship' ]]\
        .rename(columns={'authorship': 'scientificNameAuthorship'})
df_nameusage
```

```python
# get accepted name
df_nameusage = df_nameusage.merge(df_taxon[[
            'taxonID', 
            'scientificNameID' ]], how='left', left_on='acceptedNameUsageID', right_on='taxonID')\
    .merge(df_name[['ID', 'scientificName', 'authorship', 'genus', 'specificEpithet', 'infraspecificEpithet']], how='left', left_on='scientificNameID_y', right_on='ID')\
    .rename(columns={
            'taxonID_x': 'taxonID', 
            'scientificNameID_x': 'scientificNameID', 
            'scientificName_x': 'scientificName', 
            'scientificName_y': 'acceptedNameUsage',
            'authorship': 'acceptedNameAuthorship',
            'genus': 'genericName' }
        )[['taxonID',
        'taxonConceptID',
            'scientificNameID',
            'nameAccordingToID',
            'taxonomicStatus',
            'acceptedNameUsageID',
            'parentNameUsageID',
            'taxonRemarks',
            'scientificName',
            'scientificNameAuthorship',
            'acceptedNameUsage',
            'acceptedNameAuthorship',
            'genericName',
            'specificEpithet',
            'infraspecificEpithet' ]]
df_nameusage
```

```python
# get parent name
df_nameusage = df_nameusage.merge(df_taxon[['taxonID', 'scientificNameID']], how='left', left_on='parentNameUsageID', right_on='taxonID')\
    .merge(df_name[['ID', 'scientificName', 'authorship']], how='left', left_on='scientificNameID_y', right_on='ID')\
    .rename(columns={
            'taxonID_x': 'taxonID',
            'scientificNameID_x': 'scientificNameID',
            'scientificName_x': 'scientificName',
            'scientificName_y': 'parentNameUsage',
            'authorship': 'parentNameAuthorship'
        })[['taxonID',
            'taxonConceptID',
            'scientificNameID',
            'nameAccordingToID',
            'taxonomicStatus',
            'acceptedNameUsageID',
            'parentNameUsageID',
            'taxonRemarks',
            'scientificName',
            'scientificNameAuthorship',
            'acceptedNameUsage',
            'acceptedNameAuthorship',
            'genericName',
            'specificEpithet',
            'infraspecificEpithet',
            'parentNameUsage',
            'parentNameAuthorship' ]]
df_nameusage
```

```python
# add classification
df_classification = pd.read_csv('coldp/bryophytes/dwc_higherclassification.tsv', sep='\t')
df_nameusage = df_nameusage.merge(df_classification[[
        'taxonID',
        'taxonRank', 
        'kingdom', 
        'phylum', 
        'class', 
        'order', 
        'family', 
        'subfamily', 
        'genus', 
        'subgenus', 
        'higherClassification'
    ]], how='left', left_on='acceptedNameUsageID', right_on='taxonID')\
    .rename(columns={
        'taxonID_x': 'taxonID'
    })[['taxonID',
        'taxonConceptID',
        'scientificNameID',
        'nameAccordingToID',
        'taxonomicStatus',
        'acceptedNameUsageID',
        'parentNameUsageID',
        'taxonRemarks',
        'scientificName',
        'scientificNameAuthorship',
        'acceptedNameUsage',
        'acceptedNameAuthorship',
        'genericName',
        'specificEpithet',
        'infraspecificEpithet',
        'parentNameUsage',
        'parentNameAuthorship',
        'taxonRank',
        'kingdom',
        'phylum',
        'class',
        'order',
        'family',
        'subfamily',
        'genus',
        'subgenus',
        'higherClassification']]
df_nameusage
```

```python
scinames = []
accnames = []
parnames = []

for index, row in df_nameusage.iterrows():
    if isinstance(row['scientificNameAuthorship'], str):
        scinames.append(row['scientificName'] + ' ' + row['scientificNameAuthorship'])
    else:
        scinames.append(row['scientificName'])

    if isinstance(row['acceptedNameAuthorship'], str):
        accnames.append(row['acceptedNameUsage'] + ' ' + row['acceptedNameAuthorship'])
    else:
        accnames.append(row['acceptedNameUsage'])

    if isinstance(row['parentNameAuthorship'], str):
        parnames.append(row['parentNameUsage'] + ' ' + row['parentNameAuthorship'])
    else:
        parnames.append(row['parentNameUsage'])

df_nameusage['scientificName'] = scinames
df_nameusage['acceptedNameUsage'] = accnames
df_nameusage['parentNameUsage'] = parnames
df_nameusage['scientificNameAuthorship'] = df_nameusage['acceptedNameAuthorship']
df_nameusage['nomenclaturalCode'] = 'ICN'
df_nameusage['nomenclaturalStatus'] = 'legitimate'

df_nameusage = df_nameusage[[
        'taxonID',
        'scientificNameID',
        'acceptedNameUsageID',
        'parentNameUsageID',
        'nameAccordingToID',
        'taxonConceptID',
        'scientificName',
        'acceptedNameUsage',
        'parentNameUsage',
        'higherClassification',
        'kingdom',
        'phylum',
        'class',
        'order',
        'family',
        'subfamily',
        'genus',
        'specificEpithet',
        'subgenus',
        'infraspecificEpithet',
        'taxonRank',
        'scientificNameAuthorship',
        'nomenclaturalCode',
        'taxonomicStatus',
        'nomenclaturalStatus',
        'taxonRemarks' ]]

df_nameusage = df_nameusage.drop_duplicates(subset=['taxonID'])
df_nameusage.to_csv('coldp/bryophytes/dwc_taxon.tsv', sep='\t', index=False)
df_nameusage
```

```python
df_nameusage['scientificNameID'] = df_nameusage['scientificNameID'].str.replace('https://id.biodiversity.org.au/name/ausmoss/', 'ausmoss-name-')
df_nameusage.to_csv('coldp/bryophytes/dwc_taxon.tsv', sep='\t', index=False)
```
