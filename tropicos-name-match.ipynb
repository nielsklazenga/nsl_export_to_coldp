```python
import pandas as pd

df_names = pd.read_csv('data/bryophytes/names.csv')

df_trop = df_names[['scientificNameID', 'scientificName', 'canonicalName']]
df_trop.rename(columns={
    'scientificNameID': 'SourceID',
    'scientificName': 'FullNameWithAuthors',
    'canonicalName': 'FullNameNoAuthors'
}, inplace=True)

new_source_id = []
for index, row in df_trop.iterrows():
    new_source_id.append(row['SourceID'][len('https://id.biodiversity.org.au/name/ausmoss/'):])

df_trop['SourceID'] = new_source_id

df_trop.to_csv('data/bryophytes/tropicos-name-match-in.tsv', sep='\t', index=False)

df_trop
```

```python
df_trop_dat = pd.read_csv('data/bryophytes/tropicos-name-match-in_results.dat', sep='\t')
df_trop_dat
```

```python
df_altids = df_trop_dat[~df_trop_dat['OutputNameID'].isna()][['SourceID', 'OutputNameID', 'OutputTitlePageYear']]

df_altids = df_altids.astype({'OutputNameID': 'int'})
df_altids = df_altids.astype({'OutputNameID': 'str'})
df_altids = df_altids.astype({'SourceID': 'int'})
df_altids = df_altids.astype({'SourceID': 'str'})

df_altids['OutputNameID'] = 'https://tropicos.org/name/' + df_altids['OutputNameID']
df_altids['SourceID'] = 'https://id.biodiversity.org.au/name/ausmoss/' + df_altids['SourceID']


df_altids.rename(columns={'SourceID': 'ID', 'OutputNameID': 'link'}, inplace=True)

df_altids
```

```python
import os
from zipfile import ZipFile

df_nam = pd.read_csv('coldp/bryophytes/name.tsv', sep='\t')

print(df_altids.columns.to_list())
print(df_nam.columns.to_list())

df_merge = df_nam.merge(df_altids, how='left', left_on='scientificNameID', right_on='ID')

df_merge.rename(columns={'OutputTitlePageYear': 'publishedInYear', 'scientificNameID': 'ID'}, inplace=True)

print(df_merge.columns.to_list())

df_merge.to_csv('coldp/bryophytes/name.tsv', sep='\t', index=False)

os.chdir('/home/niels/code/jupyter-notebooks/nsl_export/coldp/bryophytes')

with ZipFile('nsl_bryophytes_coldp.zip', 'w') as zipobj:
    zipobj.write('taxon.tsv')
    zipobj.write('name.tsv')
    zipobj.write('synonym.tsv')
    zipobj.write('namerelation.tsv')
    zipobj.write('reference.tsv')

os.chdir('/home/niels/code/jupyter-notebooks/nsl_export')

df_merge
```
